```python
#list of moisture levels
soil_moisture=[45,60,30,55,40]
#display moisture levels
print("soil moisture levels:")
for moisture in soil_moisture:
    print(moisture,"%")
```

```python
#check fields that need irrigation
print("\nfields that need irrigation:")
for moisture in soil_moisture:
 if moisture<50:
  print(moisture,"%-irrigation needed")
 else:
  print(moisture,"%-sufficient moisture")
```

```python
crops=["rice","wheat","maize","pulses"]
yield_tons=(3.5,4.0,2.8,1.5) #tons per hectare
average_yield=sum(yield_tons)/len(yield_tons)
print("average yield:",average_yield,"tons per hectare")
```

```python
max_crop=max(zip(yield_tons,crops))
print(max_crop)
print("crop with highest yield:",max_crop[1])
```

```python
average_yield=sum(yield_tons)/len(yield_tons)
print("high yield crops:")
for i in range(len(crops)):
  if yield_tons[i]>average_yield:
    print(crops[i])
```

```python
for c,y in zip(crops,yield_tons):
  if y>average_yield:
    print(c)
```
