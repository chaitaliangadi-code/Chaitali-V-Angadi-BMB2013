```python
Yearly_Production=df.groupby("Year")["Production_Tons"].sum()
print("Yearly Production:")
Yearly_Production
```

```python
import matplotlib.pyplot as plt
plt.scatter(df["Rainfall_mm"],df["Production_Tons"])
plt.xlabel("Rainfall_mm")
plt.ylabel("Production Tons")
plt.title("Rainfall vs Production")
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.bar(Yearly_Production.index,Yearly_Production.values,color="pink")
plt.xlabel("Year")
plt.ylabel("Production Tons")
plt.title("Total Crop Production per Year")
plt.xticks(Yearly_Production.index)
plt.show()
```

```python
wheat=df[df["Crop"]=="Wheat"]
rice=df[df["Crop"]=="Rice"]
plt.figure()
plt.plot(wheat["Year"],wheat["Production_Tons"],label="Wheat")
plt.plot(rice["Year"],rice["Production_Tons"],label="Rice")
```

```python
plt.plot(wheat["Year"],wheat["Production_Tons"],label="Wheat")
plt.plot(rice["Year"],rice["Production_Tons"],label="Rice")
plt.xlabel("Year")
plt.ylabel("Production Tons")
plt.title("Crop Production Over Time")
plt.legend()
plt.show()
```

```python
data_2022=df[df["Year"]==2022]
print(data_2022)
plt.figure(figsize=(6,4))
plt.bar(data_2022["Crop"],data_2022["Production_Tons"],color=["pink","skyblue"])
plt.xlabel("Crop")
plt.ylabel("Production Tons")
plt.title("Crop Production in 2022")
plt.show()
```

```python
plt.figure(figsize=(6,4))
bars=plt.bar(Yearly_Production.index,
             Yearly_Production.values,
             color="pink",
             edgecolor="purple")
for bar in bars:
  height=bar.get_height()
  plt.text(bar.get_x()+bar.get_width()/2,
           height,
           str(height),
           ha="center",
           va="bottom")
```
