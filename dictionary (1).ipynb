```python
cgpa={1:9.1,2:8.4,3:6.5,4:6.7,5:7.8,7:9.1,8:5.1}
print(list(cgpa.keys()))
print(list(cgpa.values()))
print(list(cgpa.items()))
```

```python
keylist=list(cgpa.keys())
vallist=list(cgpa.values())
minval=min(cgpa.values())
print("value=",minval,"key",keylist[vallist.index(minval)])
```

```python
keylist=list(cgpa.keys())
vallist=list(cgpa.values())
maxval=max(cgpa.values())
print("value=",maxval,"key",keylist[vallist.index(maxval)])
```

```python
print(min(cgpa,key=cgpa.get))
print(max(cgpa,key=cgpa.get))
```
