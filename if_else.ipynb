```python
soil_moisture=[45,60,30,55,40]
print("soil moisture levels in different fields:")
for moisture in soil_moisture:
 print(moisture,"%")
```

```python
if moisture <50:
  print(moisture,"%- irrigation neede")
else:
  print(moisture ,"% -sufficient moisture")
```
