```python
print(df.head(5))
```

```python
x=df.iloc[:,0].values
y=df.iloc[:,-1].values
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=1/3,random_state=0)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

```python
print(df.shape)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
import numpy as np
x_mean=np.mean(x_train)
y_mean=np.mean(y_train)
print(x_mean)
print(y_mean)
```

```python
import pandas as pd
df=pd.read_csv("/content/Salary_Data (1).csv")
df
```

```python
#intercept and coefficient
n=len(x_train)
num,den=0,0
for i in range(n):
  num+=(x_train[i]-x_mean)*(y_train[i]-y_mean)
  den+=(x_train[i]-x_mean)**2
b1=num/den
b0=y_mean-(b1*x_mean)
print("b0:",b0)
```

```python
y_pred_code=b0+b1*x_test
print(y_pred_code)
```

```python
print(y_test)
```

```python
#mean square error
n1=len(y_test)
mse=0
for i in range(n1):
  mse+=(y_pred_code[i]-y_test[i])**2
mse=mse/n1
print(mse)
```

```python
#root mean square error
rmse=np.sqrt(mse)
print(rmse)
```
