```python
print("🌾farming game🌾")
print("let's start the game")
print("choose the option")
print("option 1.plant seeds🌱")
print("option 2.water the plants🧺")
print("option 3.harvest the plants🌿")
print("option 4.exit")
crops=0
points=0

while True:
      option=int(input("enter the option")) # Moved input inside the loop for interactivity
      if(option==1):
         print("you have chosen to plant seeds")
         crops+=1
         print("You now have crops")
      elif(option==2):
        if crops>=1:
         print("water the plants")
         points+=1 # Increment points here
         print("You watered plants and earned points",points) # Print the updated points
        else:
          print("you don't have any crops to water")
      elif(option==3):
        if crops>=1:
         print("you have chosen to harvest the plants")
         points+=crops*10
         print("Your points",crops*10)
         crops=0
         print("You now have total points.")
        else:
           print("you don't have any crops to harvest")
      elif(option==4):
         print("you have chosen to exit")
         break
      else:
          print("Invalid option. Please choose between 1 and 4.")
```
