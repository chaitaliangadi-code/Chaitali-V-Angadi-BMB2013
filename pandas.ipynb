```python
import pandas as pd
data={
    "farm":["farm A","farm B","farm C","farm D"],
    "crop":["wheat","rice","wheat","maize"],
    "area_ha":[10,15,8,12],
    "production_tons":[30,60,24,36]
}
df=pd.DataFrame(data)
print(df)
```

```python
df.head(2)
```

```python
df.tail(2    )
```

```python
df.info()
```

```python
df.describe()
```

```python
df["yield_ton_per_ha"]=df["production_tons"]/df["area_ha"]
print(df)
```

```python
print(df.iloc[[1,3]])
```

```python
df.iloc[[1,3],[0,1,2]]
```

```python
df.loc[[1,3],["farm","crop","area_ha"]]
```

```python
print(df["crop"]=="wheat")
print()
wheat_farms=df[df["crop"]=="wheat"]
print(wheat_farms)
```

```python
high_yield=df[df["yield_ton_per_ha"]>3]
print(high_yield)
```

```python
sorted_df=df.sort_values(by="production_tons",ascending=False)
print(sorted_df)
```

```python
avg_yield=df.groupby("crop")["yield_ton_per_ha"].sum()
print(avg_yield)
```
