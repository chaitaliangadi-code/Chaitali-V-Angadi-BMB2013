```python
def func(x):
  print("local x is:",x)
  x=2
  print("changed local x to:",x)
x=50
func(x)
print("x is still:",x)
```

```python
def func1():
  global y
  print("y is :",y)
  y=2
  print("changed y to:",y)
y=50
func1()
print("value of y is:",y)
```

```python
total=0
def sum(arg1,arg2):
  total=arg1+arg2
  print("inside the function local total:",total)
  return total
sum(10,20)
print("outside the function global total:",total)
```
